```python
from datetime import datetime
```

```python
# Peliculas de los ultimos 2 años
titulos[titulos.year >= datetime.now().year - 2]
```

```python
# Listar peliculas de Dracula
dracula_movies = titulos[titulos.title == "Dracula"]
dracula_movies
```

```python
# Mostrar las peliculas de Dracula desde la mas actuales a las mas antiguas
dracula_movies.sort_values("year", ascending = False)
```

```python
# Listar todas las peliculas que contengan la palabra "guerra" en su titulo
titulos[titulos.title.str.contains("guerra")].head()
```

```python
# Peliculas de Robert De Niro
elenco[elenco.name == "Robert De Niro"].head()
```

```python
# Peliculas desde el año 1994 a 2005
elenco[(elenco.year >= 1994) & (elenco.year <= 2005)]
```

```python
# Peliculas de Tom Hanks o Robert De Niro
elenco[(elenco.name == "Tom Hanks") | (elenco.name == "Robert De Niro")]
```

```python
# Nombre de actrices
elenco[elenco.type == "actress"].name
```

```python
# Cuantas peliculas estan listadas por el nombre "Dracula" ?
len(titulos[titulos.title == "Dracula"])

titulos[titulos.title == "Dracula"].shape[0]
```

```python
# Titulos mas comunes en la historia filmografica
titulos.title.value_counts().head()
```

```python
# Cual fue la primer pelicula hecha titulada "Romeo and Juliet" ?
titulos[titulos.title == "Romeo and Juliet"].sort_values(by="year").head(1)
```

```python
# Cuantas peliculas fueron hechas en el año 1950?
titulos[titulos.year == 1950].shape[0]
```

```python
# Mostrar el elenco completo de la pelicula "2001: A Space Odyssey" 
# ordenado por su clasificasión "n", ignorando los papeles que 
# no se les asigno ningun valor "n"

pelis = elenco[elenco.title == "2001: A Space Odyssey"]
pelis[pelis.n.notnull()].sort_values(by="n")
```
